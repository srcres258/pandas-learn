#### 1.2.4 任务三：判断数据是否为空，为空的地方返回True，其余地方返回False

```python
def judge_null(x: pd.Series) -> bool:
    for it in x.items():
        if pd.isna(it[1]):
            return True
    return False

df.apply(judge_null, axis=1)
```
