### 1.5 筛选的逻辑

表格数据中，最重要的一个功能就是要具有可筛选的能力，选出我所需要的信息，丢弃无用的信息。

下面我们还是用实战来学习pandas这个功能。

#### 1.5.1 任务一： 我们以"Age"为筛选条件，显示年龄在10岁以下的乘客信息。

```python
df.loc[df['Age'] < 10, :]
```

#### 1.5.2 任务二： 以"Age"为条件，将年龄在10岁以上和50岁以下的乘客信息显示出来，并将这个数据命名为midage

```python
midage = df.loc[(df['Age'] >= 10) & (df['Age'] < 50), :]
midage
```

#### 1.5.4 任务四：使用loc方法将midage的数据中第100，105，108行的"Pclass"，"Name"和"Sex"的数据显示出来

```python
lines = [99, 104, 107]
cols = ['Pclass', 'Name', 'Sex']

for line in lines:
    print(f"Line {line}:")
    datum = df.loc[line, cols]
    for col in cols:
        print(f"{col}: {datum[col]}")
    print()
```

#### 1.5.5 任务五：使用iloc方法将midage的数据中第100，105，108行的"Pclass"，"Name"和"Sex"的数据显示出来

```python
lines = [99, 104, 107]
cols = ['Pclass', 'Name', 'Sex']
col_indices = []

# Locate the index position initially.
indices = list(df.columns)
for col in cols:
    for i, index in enumerate(indices):
        if col == index:
            col_indices.append(i)
            break
assert len(cols) == len(col_indices)

for line in lines:
    print(f"Line {line}:")
    datum = df.iloc[line, col_indices]
    for col in cols:
        print(f"{col}: {datum[col]}")
    print()
```

```python
midage.iloc[[99, 104, 107], [3, 4, 5]]
```
