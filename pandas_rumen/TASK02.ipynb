#### 1.5.4 ä»»åŠ¡å››ï¼šä½¿ç”¨locæ–¹æ³•å°†midageçš„æ•°æ®ä¸­ç¬¬100ï¼Œ105ï¼Œ108è¡Œçš„"Pclass"ï¼Œ"Name"å’Œ"Sex"çš„æ•°æ®æ˜¾ç¤ºå‡ºæ¥

```python
lines = [99, 104, 107]
cols = ['Pclass', 'Name', 'Sex']

for line in lines:
    print(f"Line {line}:")
    datum = df.loc[line, cols]
    for col in cols:
        print(f"{col}: {datum[col]}")
    print()
```

#### 1.5.5 ä»»åŠ¡äº”ï¼šä½¿ç”¨ilocæ–¹æ³•å°†midageçš„æ•°æ®ä¸­ç¬¬100ï¼Œ105ï¼Œ108è¡Œçš„"Pclass"ï¼Œ"Name"å’Œ"Sex"çš„æ•°æ®æ˜¾ç¤ºå‡ºæ¥

```python
lines = [99, 104, 107]
cols = ['Pclass', 'Name', 'Sex']
col_indices = []

# Locate the index position initially.
indices = list(df.columns)
for col in cols:
    for i, index in enumerate(indices):
        if col == index:
            col_indices.append(i)
            break
assert len(cols) == len(col_indices)

for line in lines:
    print(f"Line {line}:")
    datum = df.iloc[line, col_indices]
    for col in cols:
        print(f"{col}: {datum[col]}")
    print()
```
